# Onehot encoding with many categories in a feature

```python
import pandas as pd
df = pd.read_csv('mercedez.csv', usecols = ['X0', 'X1', 'X2',  'X3', 'X4', 'X5', 'X6'])
df.head()
```

```python
for i in df.columns:
    print(df[i].value_counts())        #Getting all categorical features of all columns
```

```python
df['X0'].value_counts()    #Getting categorical features of only X0 column
```

```python
df['X0'].unique()    #Give categoical features names of column X0
```

```python
len(df['X0'].unique())
```

```python
for i in df.columns:
    print(len(df[i].value_counts()))      #Give total categories of each column one by one
```

```python
#So, when we have too many categorical values then how we use onehot encoding?
# We will apply one hot encoding on 10 most repeating categories of each column and skip the remaining features

df.X1.value_counts()    #Give numbers of each categories which means how much one category feature is repeating
```

```python
#Getting top 10 most frequent features

df.X1.value_counts().sort_values(ascending = False).head(10)
```

```python
lst_10 = df.X1.value_counts().sort_values(ascending = False).head(10).index
```

```python
lst_10 = list(lst_10)
lst_10
```

```python
import numpy as np

for features in lst_10:
    df[features] = np.where(df['X1']== features, 1, 0)
```

```python
lst_10.append('X1')
```

```python
df[lst_10]
```

# Ordinal Number encoding

```python
import datetime
```

```python
today_date = datetime.datetime.today()      #Getting todays time
today_date
```

```python
today_date - datetime.timedelta(2)         #It will subtract 2 days from todays and give us the date
```

```python
#list_compreension
days = [today_date - datetime.timedelta(x) for x in range(0, 15)]     #It will give us date of previous 15 days
```

```python
import pandas as pd
data = pd.DataFrame(days)
data.columns = ['Days']
data.head()
```

```python
data['weekday'] = data['Days'].dt.day_name()    #Getting names of week
data['weekday']
```

```python
data.head()
```

```python
dictionary = {'Monday' : 1,
    'Tuesday' : 2,
    'Wednesday' : 3,
    'Thursday' : 4,
    'Friday' : 5,
    'Saturday' : 6,
    'Sunday' : 7
}
dictionary
```

```python
data['weekday_ordinal'] = data['weekday'].map(dictionary)     
data['weekday_ordinal']
```

```python
data.head()  #So we have given a rank to each week day.
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature, ":", len(train_set[feature].unique()), 'labels')
```

### Target Guided Ordinal Encoding
#### Ordering the labels according to the taget
#### replace the labels by the join probability of being 1 or 0

```python
import pandas as pd
df = pd.read_csv('train.csv', usecols = ['Cabin', 'Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace = True)
df.head(20)
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]   #It will tell the first letter of each categoricalfeatures in cabin col
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'].value_counts()
```

```python
df.groupby(['Cabin'])['Survived'].mean()  df

#This will give the informtion that what was the percentage of survival person in cabins[A,B,C,D,E,F,G,M,T]
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
ordinal_labels2 = {k:i for i, k in enumerate(ordinal_labels, 0)}   
#As D cabin has the highest percentage of the survival so it will get higheast rank
ordinal_labels2
```

```python
df['Cabin_ordinal_labels'] = df['Cabin'].map(ordinal_labels2)
df.head()

#Now, we can drop the cabin column because we are converted into integer values
```

## Mean Encoding
#### We will replace high mean to that cabin which has highest percentage of survival

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['Mean_Ordinal'] = df['Cabin'].map(mean_ordinal)
```
